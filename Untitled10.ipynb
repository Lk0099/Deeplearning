```python
# Train and test data directories
data_train = "/content/Plant-Leaf-Disease-Detection-using-DL/train"
data_test = "/content/Plant-Leaf-Disease-Detection-using-DL/valid"

# Load the train and test data
dataset = ImageFolder(data_train, transform=transforms.Compose([
    transforms.Resize((150, 150)),
    transforms.ToTensor()
]))

test_dataset = ImageFolder(data_test, transform=transforms.Compose([
    transforms.Resize((150, 150)),
    transforms.ToTensor()
]))

# check the shape of data
img, label = dataset[0]
print(img.shape,label)

print("Follwing classes are there : \n",dataset.classes)


#plot the image
from matplotlib import pyplot as plt
def display_img(img,label):
    print(f"Label : {dataset.classes[label]}")
    plt.imshow(img.permute(1,2,0))

#display the first image in the dataset
display_img(*dataset[0])


from torch.utils.data.dataloader import DataLoader
from torch.utils.data import random_split

batch_size = 6
val_size = 2000
train_size = len(dataset) - val_size

train_data,val_data = random_split(dataset,[train_size,val_size])
print(f"Length of Train Data : {len(train_data)}")
print(f"Length of Validation Data : {len(val_data)}")

#output
#Length of Train Data : 73975
#Length of Validation Data : 2000

#load the train and validation into batches.
train_dl = DataLoader(train_data, batch_size, shuffle = True, num_workers = 4, pin_memory = True)
val_dl = DataLoader(val_data, batch_size*2, num_workers = 4, pin_memory = True)


# show train_dl k
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
from torchvision import models

def show_batch(dl):
    """Plot images grid of single batch"""
    for images, labels in dl:
        fig,ax = plt.subplots(figsize = (16,12))
        ax.set_xticks([])
        ax.set_yticks([])
        ax.imshow(make_grid(images,nrow=16).permute(1,2,0))
        break

show_batch(train_dl)





















# Define a simple CNN model for plant disease detection
class PlantDiseaseCNN(pl.LightningModule):
    def __init__(self, num_classes):
        super(PlantDiseaseCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.relu1 = nn.ReLU()
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.relu2 = nn.ReLU()
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.fc1 = nn.Linear(64 * 56 * 56, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, num_classes)

        # Initialize variables to store validation outputs
        self.validation_losses = []
        self.validation_corrects = []
        self.validation_totals = []


    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.maxpool1(x)
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.maxpool2(x)
        x = x.view(x.size(0), -1)  # Flatten
        x = self.fc1(x)
        x = self.relu3(x)
        x = self.fc2(x)
        return x

    def configure_optimizers(self):
        return optim.Adam(self.parameters(), lr=0.001)

    def training_step(self, batch, batch_idx):
        images, labels = batch
        out = self(images)                  # Generate predictions
        loss = nn.CrossEntropyLoss()(out, labels) # Calculate loss
        self.log('train_loss', loss, on_step=True, on_epoch=True) # Log training loss
        return loss

    def validation_step(self, batch, batch_idx):
        images, labels = batch
        out = self(images)
        loss = nn.CrossEntropyLoss()(out, labels)
        preds = torch.argmax(out, dim=1)
        correct = (preds == labels).sum()

        # Save validation outputs in-memory
        self.validation_losses.append(loss)
        self.validation_corrects.append(correct)
        self.validation_totals.append(len(labels))

    def on_validation_epoch_end(self):
        # Calculate and log the average validation loss and accuracy
        val_loss = torch.stack(self.validation_losses).mean()
        val_correct = torch.sum(torch.stack(self.validation_corrects))
        val_total = torch.sum(torch.stack(self.validation_totals))
        val_acc = val_correct.item() / val_total.item()

        self.log('val_loss', val_loss, on_step=False, on_epoch=True)
        self.log('val_acc', val_acc, on_step=False, on_epoch=True)

# Define dataset and dataloader (replace this with your dataset handling code)
data_transform = transforms.Compose([transforms.Resize((224, 224)),
                                     transforms.ToTensor()])
dataset = ImageFolder(root='/content/Plant-Leaf-Disease-Detection-using-DL/train', transform=data_transform)
train_loader = DataLoader(dataset, batch_size=64, shuffle=True)

# Create an instance of your LightningModule
model = PlantDiseaseCNN(num_classes=len(dataset.classes))

# Train the model
trainer = pl.Trainer(max_epochs=10)
trainer.fit(model, train_loader)

# Save the trained model in ONNX format
dummy_input = torch.randn(1, 3, 224, 224)  # Example input with dynamic batch size
onnx_path = "plant_disease_model.onnx"
dynamic_axes = {'input': {0: 'batch_size'}, 'output': {0: 'batch_size'}}
torch.onnx.export(model, dummy_input, onnx_path, verbose=True, dynamic_axes=dynamic_axes)

print(f"Model saved in ONNX format at: {onnx_path}")
```
